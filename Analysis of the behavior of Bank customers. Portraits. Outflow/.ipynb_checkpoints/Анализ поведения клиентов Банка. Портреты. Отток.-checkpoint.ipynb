## Постановка задачи и описание целей.

**Задача** - Проанализировать клиентов регионального Банка и выделить сегменты клиентов, которые склонны уходить из банка.  

**Цель исследования** - выявить набор специфических характеристик клиентов, расторгнувших контракт с Банком, и сегментировать существующих клиентов Банка с целью определения вероятности их ухода. 

**Потребитель результатов исследования** - продуктовый отдел Банка.  

**Форма представления результатов** - Отчет с результатами исследования в Jupiter, презентация полученных результатов, дашборд с основными отличиями в портретах.  

**Возможное использование результатов анализа** - Мониторинг текущих портретов клиентов по срезам моментов времени (месячным, квартальным или иным) позволит определять движение портрета от "удержанности" к "уходу", тем самым оценивать привлекательность существуемых и предлагаемых новых продуктов Банка, оценивать перспективы развития клиентской базы.

```python
def categorize_age(age):
    try:
        if 0 < age < 25:
            return '18-25'
        elif 25 <= age < 30:
            return '25-30'
        elif 30 <= age < 35:
            return '30-35'
        elif 35 <= age < 40:
            return '35-40'
        elif 40 <= age < 45:
            return '40-45'
        elif 45 <= age < 50:
            return '45-50'
        elif 50 <= age < 55:
            return '50-55'
        elif 55 <= age < 60:
            return '55-60'
        elif 60 <= age < 70:
            return '60-70'
        elif age > 70:
            return '> 70'
    except:
        pass
```

### Переименование столбцов. 

```python
# проведем переименование столбцов
data.columns = [x.lower() for x in data.columns]
data.rename(columns = {'userid':'user_id', 'creditcard':'credit_card', 'estimated_salary':'est_salary'}, inplace = True )
```

**Разбили слитное написание, заменили "верблюжий" формат на "змеиный", досократили.**

### Перевод столбца "gender" в численную бинарную форму. 

```python
# переведем столбец "gender" в численную бинарную форму
for i in range(len(data['gender'])):
    if data['gender'][i] == 'Ж':
        data['gender'][i] = 0
    elif data['gender'][i] == 'М':
        data['gender'][i] = 1
    else:
        print('ERROR! [i] = ', i, '  data["gender"][i] = ', data['gender'][i])

# изменим тип столбца "gender" на "int"
data['gender'] = data['gender'].astype('int')
```

**Для вычисления корреляции перевели столбец "gender" в численную бинарную форму.**

### Ввод дополнительных численных бинарных столбцов к столбцу "city". 

```python
# введем дополнительные столбцы бинарного представления столбца "city" 
data[['yarosl_num', 'rybinsk_num', 'rost_num']] = 0
for i in range(len(data['city'])):
    if data['city'][i] == 'Ярославль':
        data['yarosl_num'][i] = 1
    elif data['city'][i] == 'Рыбинск':
        data['rybinsk_num'][i] = 1
    elif data['city'][i] == 'Ростов Великий':
        data['rost_num'][i] = 1
    else:
        print('ERROR! [i] = ', i, '  data["city"][i] = ', data['city'][i])
```

**Аналогично для вычисления корреляции к столбцу "city" добавили три столбца с численным бинарным представлением по каждому городу.**

### Анализ пропусков. 

```python
# введем дополнительные столбцы бинарного представления столбца "balance", где 'balance_num'= 0 соответствует 
# значению NULL в столбце "balance"
data['balance_num'] = 0
for i in range(len(data['balance'])):
    if data['balance'][i] > 0:
        data['balance_num'][i] = 1
print('ПРОВЕРКА заполнения бинарного столбца "balance_num" ', data['balance'].isna().sum(), data['balance_num'].mean())

# определим набор бинарных столбцов
column_binar = ['gender', 'products', 'credit_card', 'loyalty', 'churn', 'yarosl_num', 
                'rybinsk_num', 'rost_num', 'balance_num']

# получим количественное распределение значений NULL столбца 'balance' по бинарным столбцам
display('Количественное распределение значений NULL столбца "balance" по бинарным столбцам:')
for prod in range(data['products'].nunique()):
    print('"products" = ', prod+1)
    for col in column_binar:
        try:
            print(f'{data.loc[(data["balance"].isna() & (data["products"] == prod+1)), col].sum():5.1f}  {col}') #  & (data["products"] == 2)
        except:
            pass
    print() 
```

```python
# получим сочетания значений бинарных столбцов при значении 'balance_num'= 0 и = 1        .query('balance_num == 0')
print('Сочетания значений бинарных столбцов при значении "balance_num"= 0 и 1')
display(data.pivot_table(index=['balance_num', 'products', 'credit_card', 'churn', 'loyalty'], 
                                                   values='user_id', aggfunc='count'))
```

### Портрет ушедшего и оставшегося клиента.

Результаты разведочного анализа с помощью гистограмм, полученные выше, проверим и подтвердим данными, полученными из датасета с помощью метода .describe() для исходной (initial) группы клиентов и групп оставшихся (remaining) и ушедших (goan) клиентов. Для этого:  
1. сформируем таблицу для трех групп с набором показателей датасета в натуральных показателях (чел., шт, объектов, руб.)  
2. выявим набор различных и неменяющихся показателей среди групп, выделим неменяющиеся в отдельную таблицу 
3. по таблице с отклоняющимися по группам параметрами в натуральных, абсолютных показателях составим таблицу относительных показателей, рассчитанных по базе каждой группы, т.е. отнормируем по кол-ву клиентов в каждой группе
4. выявим неизменяющиеся по группам относительные показатели и перенесем их в ранее созданную группу неменяющихся показателей  
5. создадим еще 2 вспомогательные таблицы:  
- таблицу относительных показателей групп оставшихся и ушедших клиентов, нормированных на число клиентов исходной группы (initial)  
- таблицу относительных показателей групп ушедших клиентов, нормированных на число клиентов группы оставшихся (remaining)  
6. Обнаруженные особенности распределения кол-ва продуктов Банка среди оставшихся и ушедших клиентов не могут быть получены из таблиц, производных от .describe(), поэтому эти результаты будут получены посредством сводной таблицы
7. проверим посредством сводной таблицы распределение кредитных карт среди клиентов в зависимости от кол-ва продуктов, которыми пользуется клиент
8. по итоговым таблицам в абсолютных и относительных отличающихся по группам показателей составим портреты клиента оставшегося и ушедшего

```python
#pd.reset_option('all')
display('Стат.характеристики данных всего датасета', data.drop(['user_id'],axis=1).describe(), \
        'Стат.характеристики данных в группе "оставшихся"', data.drop(['user_id'],axis=1).query('churn == 0').describe(), \
        'Стат.характеристики данных в группе "ушедших"', data.drop(['user_id'],axis=1).query('churn == 1').describe())
```

```python
# перекомпонуем информацию из таблиц .describe
# создаем пустой датафрейм, определяем столбцы
portret = pd.DataFrame(index=range(0),columns=range(14))
portret.columns = ['stat', 'n_tot', 'n_male', 'n_credit_card', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 'n_balance', 
                   'age_mean','products_mean', 'objects_mean', 'balance_mean', 'est_salary_mean']

# создаем названия портретов (профилей) выводимых в абсолютных  (кол-во человек) показателях
portret['stat'] = ['abs_val_initial', 'abs_val_remaining', 'abs_val_goan']
    
# заполняем строки
for i in range(0,3):
    if i == 0: # заполняем из основного data.describe()
        t_count = data.describe().loc['count', 'user_id']
        t_mean = data.drop(['user_id'],axis=1).describe().loc['mean', :]
    else:     # заполняем из срезов основного data.query('churn == 0 или 1').describe()   
        t_count = data.query('churn == (@i-1)').describe().loc['count', 'user_id']
        t_mean = data.query('churn == (@i-1)').drop(['user_id'],axis=1).describe().loc['mean', :]
    portret.loc[i,['n_tot', 'n_male', 'n_credit_card', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 'n_balance', \
              'age_mean','products_mean', 'objects_mean', 'balance_mean', 'est_salary_mean']] = \
    [t_count, t_mean['gender']*t_count, t_mean['credit_card']*t_count, t_mean['loyalty']*t_count, t_mean['yarosl_num']*t_count,\
    t_mean['rybinsk_num']*t_count, t_mean['rost_num']*t_count, t_mean['balance_num']*t_count, \
    t_mean['age'], t_mean['products'], t_mean['objects'], t_mean['balance'], t_mean['est_salary']]
    
# выведем таблицу portret, содержащую набор количественных значений начальгой группы клиентов, оставшихся и ушедших 
# в абсолютных показателях
display('ПУНКТ 1: Портреты (профили) исходной группы клиентов initial, группы оставшихся remaining и группы ушедших goan \
в абсолютных показателях', portret)   

#pd.set_option('display.float_format', '{:.2f}'.format)

# выведем набор близких показателей в отдельную таблицу "нет различий"
portret_not_diff = portret.drop(['n_tot', 'n_male', 'n_credit_card', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 
                                 'n_balance', 'age_mean'], axis=1)

# выведем таблицу "нет различий"
display('ПУНКТ 2: Показатели портретов (профилей) исходной группы клиентов, оставшихся и ушедших, практически идентичные', 
       portret_not_diff)   

# удалим из таблицы портретов набор близких показателей 
portret = portret.drop(['products_mean', 'objects_mean', 'balance_mean', 'est_salary_mean'], axis=1)

# создадим таблицу портретов в относительных единицах по отношению к собственной базе
# создаем пустой датафрейм, определяем столбцы
portret_rel_own_base = pd.DataFrame(index=range(0),columns=range(9))
portret_rel_own_base.columns = ['stat','n_tot','n_male','n_credit_card','n_loyalty','n_yarosl','n_rybinsk','n_rost',
                                'n_balance']
# создаем названия портретов (профилей) в относительных единицах (%) по отношению к собственной базе
portret_rel_own_base['stat'] = ['rel_own_base_initial', 'rel_own_base_remaining', 'rel_own_base_goan']
    
columns_n_value = ['n_tot', 'n_male', 'n_credit_card', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 'n_balance']
for row in range(0,3):    # перебираем строки
    for col in (columns_n_value):        # формируем значения ячеек в каждой строке, делим кол-во клиентов 
                                         # в этом показателе на полное кол-во клиентов в этой группе
        portret_rel_own_base.loc[row,col] = portret.loc[row,col]*100.0/portret.loc[row, 'n_tot']

# переименовываем столбцы
portret_rel_own_base = portret_rel_own_base.rename(columns={'n_tot':'%_tot', 'n_male':'%_male', 
                                                            'n_credit_card':'%_credit_card','n_loyalty':'%_loyalty', 
                                                            'n_yarosl':'%_yarosl', 'n_rybinsk':'%_rybinsk', 
                                                            'n_rost':'%_rost', 'n_balance':'%_balance'})        
 
display('ПУНКТ 3: Относительные показатели (%) по отношению к собственной базе портретов (профилей) исходной группы клиентов, \
оставшихся и ушедших', portret_rel_own_base)   
# перересем столбец n_credit_card из таблицы portret и столбец %_credit_card из таблицы portret_rel_own_base 
# в таблицу portret_not_diff
portret_not_diff['n_credit_card'] = portret['n_credit_card']
portret_not_diff['%_credit_card'] = portret_rel_own_base['%_credit_card']

portret = portret.drop(['n_credit_card'],axis=1)
portret_rel_own_base = portret_rel_own_base.drop(['%_credit_card'],axis=1)

display('ПУНКТ 4: Показатели портретов (профилей) исходной группы клиентов, оставшихся и ушедших, практически идентичные', 
        portret_not_diff, 
        'Портреты (профили) исходной группы клиентов initial, группы оставшихся remaining и группы ушедших goan \
в абсолютных показателях', portret, 
        'Относительные показатели (%) по отношению к собственной базе портретов (профилей) исходной группы клиентов, \
оставшихся и ушедших', portret_rel_own_base)

# создадим таблицу портретов в относительных единицах по отношению к базе исходной группы
# создаем пустой датафрейм, определяем столбцы
portret_rel_base_init = pd.DataFrame(index=range(0),columns=range(8))
portret_rel_base_init.columns = ['stat','n_tot','n_male','n_loyalty','n_yarosl','n_rybinsk','n_rost',
                                'n_balance']
portret_rel_base_init['stat'] = ['rel_base_init_remaining', 'rel_base_init_goan']
    
columns_n_value = ['n_tot', 'n_male', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 'n_balance']
for row in range(1,3):       # перебираем строки 
    for col in (columns_n_value):        # формируем значения ячеек в каждой строке, делим кол-во клиентов 
                                         # в этом показателе на кол-во клиентов в этом показателе в исходной  группе
        portret_rel_base_init.loc[row-1,col] = portret.loc[row,col]*100.0/portret.loc[0, col]

portret_rel_base_init = portret_rel_base_init.rename(columns={'n_tot':'%_tot', 'n_male':'%_male', 
                                                            'n_loyalty':'%_loyalty', 
                                                            'n_yarosl':'%_yarosl', 'n_rybinsk':'%_rybinsk', 
                                                            'n_rost':'%_rost', 'n_balance':'%_balance'})   

display('ПУНКТ 5 (1): Относительные показатели (%) портретов (профилей) оставшихся и ушедших к базе исходной группы ', portret_rel_base_init)   

# создадим таблицу портретов в относительных единицах по отношению к базе группы оставшихся
# создаем пустой датафрейм, определяем столбцы
portret_rel_base_remain = pd.DataFrame(index=range(0),columns=range(8))
portret_rel_base_remain.columns = ['stat','n_tot','n_male','n_loyalty','n_yarosl','n_rybinsk','n_rost',
                                'n_balance']
portret_rel_base_remain['stat'] = ['rel_base_remain_goan']
    
columns_n_value = ['n_tot', 'n_male', 'n_loyalty', 'n_yarosl', 'n_rybinsk', 'n_rost', 'n_balance']
for col in (columns_n_value):
    portret_rel_base_remain.loc[0,col] = portret.loc[2,col]*100.0/portret.loc[1, col]

portret_rel_base_remain = portret_rel_base_remain.rename(columns={'n_tot':'%_tot', 'n_male':'%_male', 
                                                            'n_loyalty':'%_loyalty', 
                                                            'n_yarosl':'%_yarosl', 'n_rybinsk':'%_rybinsk', 
                                                            'n_rost':'%_rost', 'n_balance':'%_balance'})  

display('ПУНКТ 5 (2): Относительные показатели (%) портретов (профилей) группы ушедших к базе группы оставшихся',
        portret_rel_base_remain)

# создадим таблицу распределения по кол-ву продуктов Банка у клиентов оставшейся и ушедшей групп
prod_churn_distr = data.pivot_table(index='products', columns='churn', values='user_id', aggfunc='count')\
                        .reset_index().rename(columns={0:'n_prod_rem', 1:'n_prod_goan'}).fillna(0)
prod_churn_distr['%_prod_rem_own'] = prod_churn_distr['n_prod_rem']*100.0/prod_churn_distr['n_prod_rem'].sum()
prod_churn_distr['%_prod_goan_own'] = prod_churn_distr['n_prod_goan']*100.0/prod_churn_distr['n_prod_goan'].sum()
for i in range(0,4):
    prod_churn_distr.loc[i,'%_prod_rem_init'] = prod_churn_distr.loc[i,'n_prod_rem']*100.0\
    /(prod_churn_distr.loc[i,'n_prod_rem'] + prod_churn_distr.loc[i,'n_prod_goan'])
    prod_churn_distr.loc[i,'%_prod_goan_init'] = prod_churn_distr.loc[i,'n_prod_goan']*100.0\
    /(prod_churn_distr.loc[i,'n_prod_rem'] + prod_churn_distr.loc[i,'n_prod_goan'])

display('ПУНКТ 6: Распределение по кол-ву продуктов Банка у клиентов оставшейся и ушедшей групп', prod_churn_distr)

prod_cred_distr = data.pivot_table(index='products', columns='credit_card', values='user_id', aggfunc='count')\
                        .reset_index().rename(columns={0:'n_without', 1:'n_with'}).fillna(0)
prod_cred_distr['%_n_without'] = prod_cred_distr['n_without']*100.0/prod_cred_distr['n_without'].sum()
prod_cred_distr['%_n_with'] = prod_cred_distr['n_with']*100.0/prod_cred_distr['n_with'].sum()
for i in range(0,4):
    prod_cred_distr.loc[i,'%_without'] = prod_cred_distr.loc[i,'n_without']*100.0\
    /(prod_cred_distr.loc[i,'n_without'] + prod_cred_distr.loc[i,'n_with'])
    prod_cred_distr.loc[i,'%_with'] = prod_cred_distr.loc[i,'n_with']*100.0\
    /(prod_cred_distr.loc[i,'n_without'] + prod_cred_distr.loc[i,'n_with'])
display('ПУНКТ 7: Распределение кредитных карт в зависимости от кол-ва продуктов Банка, используемым клиентом', 
        prod_cred_distr)
for i in range(1,5):
    print('Доля клиентов с кол-вом продуктов Банка ',i,' среди оставшихся клиентов:', 
          round(data.query('(churn == 0) & (products == @i)')['user_id'].count()*100.0/\
                    data.query('churn == 0')['user_id'].count(), 2),'%')
```

## Гипотезы: формулировка и проверка.

### Гипотеза различия дохода между оставшимися и ушедшими клиентами.

**H_0:** Средний доход "est_salary" ушедших (отточных) клиентов Банка РАВЕН среднему доходу "est_salary" оставшихся (удержанных) клиентов Банка.  
**Н_1:** Средний доход "est_salary" ушедших (отточных) клиентов Банка НЕ РАВЕН среднему доходу "est_salary" оставшихся (удержанных) клиентов Банка. 

<div class="alert alert-block alert-success">
<b>✔️ Комментарий ревьюера:</b>  Гипотеза сформулирована
</div>

**Обоснование метода:** Даны 2 выборки, первая - клиенты Банка, не расторгавшие договор с Банком, размером 7963 записи, и вторая - клиенты Бака, расторгнувшие договор с Банком, размером 2037 записей. По этим выборкам надо оценить равенство средних значений соответствующих этим выборкам генеральных совокупностей. В данном случае проверяются средние значения переменной "est_salary" - зарплаты клиентов. В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: **scipy.stats.ttest_ind (array1, array2, equal_var),** где array1, array2 - выборки, equal_var специальный параметр, который мы выбираем равным False, поскольку выборки разного размера.  
Уровень значимости выбираем 5%.

```python
# вызов метода для проверки гипотезы
results = st.ttest_ind(data.query('churn == 0')['est_salary'], data.query('churn == 1')['est_salary'], equal_var = False) 
# задаем значение уровня значимости
alpha = 0.05 
# выводим значение p-value на экран 
print('p-value = ', results.pvalue)
if results.pvalue < alpha: # условный оператор с выводом строки с ответом
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Не получилось отвергнуть нулевую гипотезу  
H_0: Средний доход "est_salary" ушедших (отточных) клиентов Банка РАВЕН среднему доходу "est_salary" оставшихся (удержанных) клиентов Банка.**

### Гипотеза различия остатка баланса на счете между оставшимися и ушедшими клиентами.

**H_0:** Средний остаток баланса на счете "balance" ушедших (отточных) клиентов Банка РАВЕН среднему остатку баланса на счете "balance" оставшихся (удержанных) клиентов Банка.  
**Н_1:** Средний остаток баланса на счете "balance" ушедших (отточных) клиентов Банка НЕ РАВЕН среднему остатку баланса на счете "balance" оставшихся (удержанных) клиентов Банка.  

```python
# вызов метода для проверки гипотезы
results = st.ttest_ind(data.query('(churn == 0) & (balance > 0)')['balance'], 
                       data.query('(churn == 1) & (balance > 0)')['balance'], equal_var = False) 
# задаем значение уровня значимости
alpha = 0.05 
# выводим значение p-value на экран 
print('p-value = ', results.pvalue)
if results.pvalue < alpha: # условный оператор с выводом строки с ответом
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### z-test

**H_0:** Доля мужчин в группе ушедших (отточных) клиентов Банка РАВНА доле мужчин в группе оставшихся (удержанных) клиентов Банка.  
**Н_1:** Доля мужчин в группе ушедших (отточных) клиентов Банка НЕ РАВНА И МЕНЬШЕ доли мужчин в группе оставшихся (удержанных) клиентов Банка. 

```python
# критический уровень статистической значимости
alpha = .05 
successes = np.array([data.query('(churn == 1) & (gender == 1)')['gender'].count(), 
                      data.query('(churn == 0) & (gender == 1)')['gender'].count()])
trials = np.array([data.query('churn == 1')['gender'].count(), 
                   data.query('churn == 0')['gender'].count()])

display('successes', successes, 'trials', trials)
# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

display('p_combined', p_combined)

# разница пропорций в датасетах
difference = p1 - p2

display('ifference', difference)

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

display('z_value', z_value)

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  

p_value = distr.cdf(z_value)   #   (1 - distr.cdf(abs(z_value))) * 2 

display(distr.cdf(abs(z_value)), 'p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    ) 
```

**Таким образом,**  в результате проведения z-теста нулевая гипотеза Н_0 о равенстве долей мужчин в выборках оставшихся и ушедших клиентов отвергнута в пользу альтернативной гипотезы  
**Н_1: Доля мужчин в группе ушедших (отточных) клиентов Банка статистически значимо НЕ РАВНА И МЕНЬШЕ доли мужчин в группе оставшихся (удержанных) клиентов Банка.**
