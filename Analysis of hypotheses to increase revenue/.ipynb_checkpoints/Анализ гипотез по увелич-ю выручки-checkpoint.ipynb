### –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –¥–∞–Ω–Ω—ã—Ö ###

```python
orders = pd.read_csv('https://code.s3.yandex.net/datasets/orders.csv') # –∑–∞–≥—Ä—É–∑–∏–º csv-—Ñ–∞–π–ª –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤—Ç–æ—Ä–æ–π —á–∞—Å—Ç–∏
visitors = pd.read_csv('https://code.s3.yandex.net/datasets/visitors.csv') # –∑–∞–≥—Ä—É–∑–∏–º csv-—Ñ–∞–π–ª –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤—Ç–æ—Ä–æ–π —á–∞—Å—Ç–∏
# –ø–æ—Å–º–æ—Ç—Ä–∏–º –æ–±—â—É—é –∏–Ω—Ñ-—Ü–∏—é –ø–æ —ç—Ç–∏–º —Ñ–∞–π–ª–∞–º
print('--'*10, 'dataset "orders" –∏ –µ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã', '--'*10)
display(orders.shape)
orders.info()
display(orders.head())
display(orders.columns)

print('orders["date"].min =', orders['date'].min(), ';  orders["date"].max =', orders['date'].max())
print('orders["revenue"].min =', orders['revenue'].min(), ';  orders["revenue"].max =', orders['revenue'].max())
print('orders["group"].unique():', orders['group'].unique())
print('–ö–æ–ª-–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–µ–π =', orders.duplicated().sum(), '\n')

# –ø–æ—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í
print('–û–±—â–µ–µ –∫–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π:', orders['visitorId'].count())
print('–ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –ê:', orders[orders['group'] == 'A']['visitorId'].count())
print('–ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –í:', orders[orders['group'] == 'B']['visitorId'].count(), '\n')

print('--'*10, 'dataset "visitors" –∏ –µ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã', '--'*10)
print(visitors.shape)
visitors.info()
display(visitors.head())
print(visitors.columns)
print('visitors["date"].min =', visitors['date'].min(), ';  visitors["date"].max =', visitors['date'].max())
print('visitors["visitors"].min =', visitors['visitors'].min(), ';  visitors["visitors"].max =', visitors['visitors'].max())
print('visitors["group"].unique():', visitors['group'].unique())
print('–ö–æ–ª-–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–µ–π =', visitors.duplicated().sum(), '\n')
# –ø–æ—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í
print('–û–±—â–µ–µ –∫–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π:', visitors['visitors'].sum())
print('–ö–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –ê:', visitors[visitors['group'] == 'A']['visitors'].sum())
print('–ö–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –í:', visitors[visitors['group'] == 'B']['visitors'].sum(), '\n')

# –ø—Ä–∏–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞—Ç –∫ —Ç–∏–ø—É "datetime64"
orders['date'] = orders['date'].astype('datetime64')
visitors['date'] = visitors['date'].astype('datetime64')
print('–î–∞—Ç—ã –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö orders["date"], visitors["date"] –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –∫ —Ç–∏–ø—É:', orders['date'].dtype, visitors['date'].dtype)
```

**–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í –¥–æ–≤–æ–ª—å–Ω–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ, —Ä–∞–∑–ª–∏—á–∏–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –º–µ–Ω–µ–µ 0,5% (0,478%).**

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –ó–∞–≥—Ä—É–∑–∏–ª–∏ –¥–∞–Ω–Ω—ã–µ –æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è—Ö –∏ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è—Ö, –≤—ã–ø–æ–ª–Ω–∏–ª–∏ –∞–Ω–∞–ª–∏–∑ –∏ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫—É - –æ—Ç–ª–∏—á–Ω–æ!

### –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤—ã–±–æ—Ä–æ–∫ –≥—Ä—É–ø–ø –ê –∏ –í –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ. –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π.

```python
# –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–∞–ª–∏—á–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í
visitorId_orders_A = orders[orders['group'] == 'A']['visitorId']
visitorId_orders_B = orders[orders['group'] == 'B']['visitorId']
visitorId_orders_AB = orders.query('(visitorId in @visitorId_orders_A) & (visitorId in @visitorId_orders_B)')['visitorId']
print('visitorId, –ø–æ–ø–∞–≤—à–∏–µ –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã:', len(visitorId_orders_AB.unique()),'\n')

# —Å–æ–∑–¥–∞–¥–∏–º —Å—Ä–µ–∑ "orders" –ø–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è–º, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í
orders_intersect = orders.query('visitorId in @visitorId_orders_AB')[['transactionId', 'visitorId', 'date', 'revenue', 'group']]
display('orders_intersect', orders_intersect)

# –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤ —É –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í
display('–ü–µ—Ä–µ—Å–µ–∫–∞—é—â–∏–µ—Å—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ - –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤', orders_intersect.groupby('visitorId', as_index=False)
        .agg({'transactionId':'count', 'revenue':'sum'})
        .sort_values(by='transactionId', ascending=False))
# –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Å—É–º–º—ã –∑–∞–∫–∞–∑–æ–≤ —É –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í
display('–ü–µ—Ä–µ—Å–µ–∫–∞—é—â–∏–µ—Å—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ - —Å—É–º–º–∞ –∑–∞–∫–∞–∑–æ–≤', orders_intersect.groupby('visitorId', as_index=False)
      .agg({'transactionId':'count', 'revenue':'sum'})
      .sort_values(by='revenue', ascending=False))
# –ø–æ—Å–º–æ—Ç—Ä–∏–º –¥–æ–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í, –≤ —á–∏—Å–ª–µ–Ω–Ω–æ–º —Å–æ—Å—Ç–∞–≤–µ, –∫–æ–ª-–≤–µ –∑–∞–∫–∞–∑–æ–≤ –∏ —Å—É–º–º–∞—Ö –∑–∞–∫–∞–∑–æ–≤
print('{0:.2%}'.format(len(orders_intersect['visitorId'].unique())/len(orders['visitorId'].unique())),\
      '- –¥–æ–ª—è –ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ —Å–æ—Å—Ç–∞–≤–µ')
print('{0:.2%}'.format(orders_intersect['transactionId'].count()/orders['transactionId'].count()),
'- –¥–æ–ª—è –ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –∫–æ–ª-–≤–µ –∑–∞–∫–∞–∑–æ–≤')
print('{0:.2%}'.format(orders_intersect['revenue'].sum()/orders['revenue'].sum()),
      '- –¥–æ–ª—è –ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ —Å—É–º–º–∞—Ö –∑–∞–∫–∞–∑–æ–≤ \n')

# —Å–æ–∑–¥–∞–¥–∏–º —Ñ—Ä–µ–π–º —Å –º—É–ª—å—Ç–∏–Ω–¥–µ–∫—Å–æ–º 'date', 'group' –∏ –∫–æ–ª-–≤–æ–º –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í, –Ω–∞ —ç—Ç–∏ –¥–∞—Ç—É –∏ –≥—Ä—É–ø–ø—É
dates_group_N_intersect = orders_intersect.groupby(['date', 'group'])['visitorId'].count()
# –ø—Ä–∏–≤–µ–¥–µ–º –¥–∞—Ç–∞ —Ñ—Ä–µ–π–º "visitors" –∫ —Ç–∞–∫–æ–º—É –∂–µ –≤–∏–¥—É —Å –º—É–ª—å—Ç–∏–Ω–¥–µ–∫—Å–æ–º 'date', 'group'
visitors_dates_group = visitors.groupby(['date', 'group'])['visitors'].sum()

# –ø–æ–ª—É—á–∏–º –Ω–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É "visitors" —Å –≤—ã—á—Ç–µ–Ω–Ω—ã–º–∏ –∫–æ–ª-–≤–∞–º–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í, –ø–æ –∫–∞–∂–¥–æ–π –¥–∞—Ç–µ –∏ –≥—Ä—É–ø–ø–µ
visitors_new = visitors_dates_group.sub(dates_group_N_intersect, fill_value=0).reset_index()
visitors_new.columns = ['date', 'group', 'visitors'] 
print(visitors_new)

# —Å–æ—Ö—Ä–∞–Ω–∏–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞—Ç–∞—Å–µ—Ç—ã "orders" –∏ "visitors" –≤ –∫–æ–ø–∏—è—Ö
orders_copy = orders
visitors_copy = visitors
# —É–¥–∞–ª–∏–º –∏–∑ "orders" –∏ "visitors" –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã, –∫–∞–∫ –æ—à–∏–±–æ—á–Ω—ã–µ
orders = orders.query('visitorId not in @visitorId_orders_AB')
visitors = visitors_new

# –ø–æ—Å–º–æ—Ç—Ä–∏–º –æ–±—â—É—é –∏–Ω—Ñ-—Ü–∏—é –ø–æ –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–º—É —Ñ–∞–π–ª—É "orders"
print('--'*10, '–æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–π "orders" (–±–µ–∑ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã) –∏ –µ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã', '--'*10)
display(orders.shape)
orders.info()
print('orders["date"].min =', orders['date'].min(), ';  orders["date"].max =', orders['date'].max())
print('orders["revenue"].min =', orders['revenue'].min(), ';  orders["revenue"].max =', orders['revenue'].max())
print('orders["group"].unique():', orders['group'].unique())
print('–ö–æ–ª-–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–µ–π =', orders.duplicated().sum(), '\n')
# –ø–æ—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í
print('–û–±—â–µ–µ –∫–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π:', orders['visitorId'].count())
print('–ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –ê:', orders[orders['group'] == 'A']['visitorId'].count())
print('–ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –í:', orders[orders['group'] == 'B']['visitorId'].count(), '\n')
# –ø—Ä–æ–≤–µ—Ä–∏–º –Ω–∞–ª–∏—á–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã –ê –∏ –í
visitorId_orders_A_2 = orders[orders['group'] == 'A']['visitorId']
visitorId_orders_B_2 = orders[orders['group'] == 'B']['visitorId']
visitorId_orders_AB_2 = orders.query('(visitorId in @visitorId_orders_A_2) & (visitorId in @visitorId_orders_B_2)')['visitorId']
print('visitorId, –ø–æ–ø–∞–≤—à–∏–µ –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã:', visitorId_orders_AB_2.count(),'\n')

# –ø–æ—Å–º–æ—Ç—Ä–∏–º –æ–±—â—É—é –∏–Ω—Ñ-—Ü–∏—é –ø–æ –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–æ–º—É —Ñ–∞–π–ª—É "visitors"
print('--'*10, '–æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–π "visitors" (–±–µ–∑ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –≤–æ—à–µ–¥—à–∏—Ö –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã) –∏ –µ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã', '--'*10)
print('–û–±—â–µ–µ –∫–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π:', visitors['visitors'].sum())
print('–ö–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –ê:', visitors[visitors['group'] == 'A']['visitors'].sum())
print('–ö–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ –í:', visitors[visitors['group'] == 'B']['visitors'].sum(), '\n')
    
```

### –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º. –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
# —Å—Ç—Ä–æ–∏–º –º–∞—Å—Å–∏–≤ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø
dates_groups = orders[['date', 'group']].drop_duplicates().sort_values(by=['date', 'group'])

# –∞–≥—Ä–µ–≥–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É "orders"
orders_agregg = dates_groups.apply(
    lambda x:orders[
        np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])
                    ]
        .agg({'date':'max', 'group':'max', 'visitorId':'nunique', 'transactionId':'nunique', 'revenue':'sum'}), axis=1
                                    )\
        .sort_values(by=['date', 'group'])
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü–∞–º –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ —Ç–∞–±–ª–∏—Ü–µ "orders_agregg"
orders_agregg.columns = ['date', 'group', 'cum_N_buyers', 'cum_N_orders', 'cum_revenue']  

display('–ê–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ orders_agregg:', orders_agregg.head())
orders_agregg.info()

# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_agregg_A = orders_agregg[orders_agregg['group'] == 'A']
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_agregg_B = orders_agregg[orders_agregg['group'] == 'B']

print('–ö–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ orders_agregg_A:', orders_agregg_A['group'].count())
print('–ö–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ orders_agregg_B:', orders_agregg_B['group'].count())

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –ê
plt.plot(orders_agregg_A['date'], orders_agregg_A['cum_revenue'], label='A')
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –í
plt.plot(orders_agregg_B['date'], orders_agregg_B['cum_revenue'], label='B')
plt.legend()
plt.title('–ì—Ä–∞—Ñ–∏–∫–∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=18, color='blue')
plt.ylabel('–í—ã—Ä—É—á–∫–∞, $*100000', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.show()
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_agregg_–ê–í" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_agregg_–ê" –∏ "orders_agregg_B"
orders_agregg_AB = orders_agregg_A.merge(orders_agregg_B, left_on='date', right_on='date', how='left', suffixes=['A','B'])

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A
plt.plot(orders_agregg_AB['date'], (orders_agregg_AB['cum_revenueB']/orders_agregg_AB['cum_N_ordersB'])\
         /(orders_agregg_AB['cum_revenueA']/orders_agregg_AB['cum_N_ordersA']) - 1, label='(B/A - 1)')
plt.legend()
plt.title('–ì—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A', fontsize=18, color='blue')
#plt.ylabel('–í—ã—Ä—É—á–∫–∞, $', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
#plt.axhline(y=0.2, color='green', linestyle='--')
plt.axhline(y=0, color='black', linestyle='-')
plt.show()
```

```python
# –∞–≥—Ä–µ–≥–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É "visitors"
visitors_agregg = dates_groups.apply(
    lambda x:visitors[
        np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])
                    ]
        .agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1
                                    )\
        .sort_values(by=['date', 'group'])
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü–∞–º –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ —Ç–∞–±–ª–∏—Ü–µ "visitors_agregg"
visitors_agregg.columns = ['date', 'group', 'cum_N_visitors']  

# –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
display(visitors_agregg.head())
visitors_agregg.info()

# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_agregg" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_agregg" –∏ "visitors_agregg"
orders_visitors_agregg = orders_agregg.merge(visitors_agregg, left_on=['date', 'group'], right_on=['date', 'group'], how='left')

# –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
display(orders_visitors_agregg.head())
orders_visitors_agregg.info()

# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_visitors_agregg_A = orders_visitors_agregg[orders_visitors_agregg['group'] == 'A']
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_visitors_agregg_B = orders_visitors_agregg[orders_visitors_agregg['group'] == 'B']

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –ê
plt.plot(orders_visitors_agregg_A['date'], orders_visitors_agregg_A['cum_N_orders']/orders_visitors_agregg_A['cum_N_visitors'],\
         label='A')
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –í
plt.plot(orders_visitors_agregg_B['date'], orders_visitors_agregg_B['cum_N_orders']/orders_visitors_agregg_B['cum_N_visitors'],\
         label='B')
plt.legend()
plt.title('–ì—Ä–∞—Ñ–∏–∫–∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=18, color='blue')
plt.ylabel('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.axhline(y=0.0290, color='red', linestyle='dotted')
plt.axhline(y=0.0255, color='green', linestyle='dotted')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.show()
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_visitors_agregg_–ê–í" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_visitors_agregg_–ê" –∏ "orders_visitors_agregg_B"
orders_visitors_agregg_AB = orders_visitors_agregg_A.merge(orders_visitors_agregg_B, left_on='date', right_on='date', \
                                                           how='left', suffixes=['A','B'])

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A
plt.plot(orders_visitors_agregg_AB['date'], (orders_visitors_agregg_AB['cum_N_ordersB']/orders_visitors_agregg_AB['cum_N_visitorsB'])\
         /(orders_visitors_agregg_AB['cum_N_ordersA']/orders_visitors_agregg_AB['cum_N_visitorsA']) - 1, label='(B/A - 1)')
plt.legend()
plt.title('–ì—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A', \
          fontsize=18, color='blue')
#plt.ylabel('–í—ã—Ä—É—á–∫–∞, $', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.axhline(y=0.14, color='green', linestyle='--')
plt.axhline(y=0, color='black', linestyle='-')
plt.show()
```

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –£—á—Ç–µ–Ω–æ. –ú–∞–ª–æ –∫–æ–≥–¥–∞ –ø—Ä–∏–º–µ–Ω—è–ª –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ —ç—Ç–æ—Ç –∫—Ä–∏—Ç–µ—Ä–∏–π - –æ–±—ã—á–Ω–æ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–ª —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±—É—Ç—Å—Ç—Ä–∞–ø–æ–º –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª `ttest`. –ü–æ –ø–æ–≤–æ–¥—É –Ω–µ—Å–≤—è–∑–Ω–æ—Å—Ç–∏ - –∞–±—Å–æ–ª—é—Ç–Ω—É—é –º–∞–ª–æ –∫–æ–≥–¥–∞ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏, –µ—Å–ª–∏ –ø—Ä—è–º —á–µ—Å—Ç–Ω–æ-—á–µ—Å—Ç–Ω–æ –Ω–µ—Å–≤—è–∑–Ω–æ—Å—Ç—å —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å. –ü–æ—ç—Ç–æ–º—É –æ–±—ã—á–Ω–æ –µ—Å—Ç—å –∫–∞–∫–∏–µ-—Ç–æ –¥–æ–ø—É—â–µ–Ω–∏—è. –ö–∞–∫–∏–µ - –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞, –≤—Å—ë –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª-–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ –ê
visitors_transactions_A = orders[orders['group'] == 'A'].groupby('visitorId', as_index=False)['transactionId'].count()
visitors_transactions_A.columns = ['visitorId', 'N_transactions']
print('–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è visitors_transactions_A:', sorted(visitors_transactions_A['N_transactions'].unique()))

# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª-–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ B
visitors_transactions_B = orders[orders['group'] == 'B'].groupby('visitorId', as_index=False)['transactionId'].count()
visitors_transactions_B.columns = ['visitorId', 'N_transactions']
print('–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è visitors_transactions_B:', sorted(visitors_transactions_B['N_transactions'].unique()))

display('visitors_transactions_A:', visitors_transactions_A, 'visitors_transactions_B:', visitors_transactions_B)

# ------------------------------------------------------------------------
print('--'*10, '–ì–ò–°–¢–û–ì–†–ê–ú–ú–´ "visitors_transactions_A" –∏ "visitors"')
plt.hist(visitors_transactions_A['N_transactions'], bins=3, color='gray', label='visitors_transactions_A')
plt.legend()
plt.show()
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏—è –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞: ', round(st.shapiro(visitors_transactions_A['N_transactions'])[1], 42))

plt.hist(visitors.groupby('date')['visitors'].sum(), bins=7, color='gray', label='visitors')
plt.legend()
plt.show()
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏—è –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞: ', round(st.shapiro(visitors.groupby('date')['visitors'].sum())[1], 3), '\n')
# -----------------------------------------------------------------------------

# –ø–æ–¥—Å—á–∏—Ç–∞–µ–º –∫–æ–ª-–≤–æ –Ω—É–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –¥–ª—è —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –≤—ã–±–æ—Ä–∫–∏ –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏
t1 = len(visitors_transactions_A['visitorId'])
t2 = visitors[visitors['group'] == 'A']['visitors'].sum()
N_zero_A = t2 - t1
print(' t1 - –ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ;', '\n',
      't2 - –û–±—â–µ–µ –∫–æ–ª-–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≥—Ä—É–ø–ø–µ;', '\n', 
      'N_zero_ - –ö–æ–ª-–≤–æ –Ω—É–ª–µ–π, –¥–æ–±–∞–≤–ª—è–µ–º–æ–µ –≤ –≤—ã–±–æ—Ä–∫—É –≥—Ä—É–ø–ø—ã;', '\n')
print('t1 =', t1, '   t2 =', t2, '   N_zero_A =', N_zero_A)

t1 = len(visitors_transactions_B['visitorId'])
t2 = visitors[visitors['group'] == 'B']['visitors'].sum()
N_zero_B = t2 - t1
print('t1 =', t1, '   t2 =', t2, '   N_zero_B =', N_zero_B)

# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –ø–æ –≥—Ä—É–ø–ø–µ –ê –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏
sample_A = pd.concat([visitors_transactions_A['N_transactions'],pd.Series(0,index=np.arange(N_zero_A), \
                                                                         name='transactions')],axis=0)
# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –ø–æ –≥—Ä—É–ø–ø–µ –í –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏
sample_B = pd.concat([visitors_transactions_B['N_transactions'],pd.Series(0,index=np.arange(N_zero_B), \
                                                                         name='transactions')],axis=0)
display(sample_A, sample_B)

print('p-value = ', round(st.mannwhitneyu(sample_A, sample_B)[1], 5))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B =','{0:.2%}'.format(sample_B.mean()/sample_A.mean()- 1))

         
```

**–ü–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º –ø–æ–ª—É—á–∞–µ–º —Å–ª–µ–¥—É—é—â—É—é –æ—Ü–µ–Ω–∫—É —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –Ω–∞–ª–∏—á–∏—è —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –í –∏ –ê.**  

**–ü—Ä–∏ –ø–æ—Ä–æ–≥–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∞ = 0,05 –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–∏—Å–ª–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í –Ω–µ—Ç, - –æ—Ç–≤–µ—Ä–≥–∞–µ–º. –í—ã–±–∏—Ä–∞–µ–º –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–∏—Å–ª–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ B –∏ –ê –µ—Å—Ç—å. –ü—Ä–∏ —ç—Ç–æ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ 16%.**

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b> –û—Ü–µ–Ω–∏–ª–∏ –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–∞ —Å—ã—Ä—ã—Ö –¥–∞–Ω–Ω—ã—Ö - –æ—Ç–ª–∏—á–Ω–æ!

### –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º. –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

**–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã.  
–ù—É–ª–µ–≤–∞—è: —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ B –∏ –ê –Ω–µ—Ç.  
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ B –∏ –ê –µ—Å—Ç—å.**

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ –ê
visitors_revenue_A = orders[orders['group'] == 'A'].groupby('visitorId', as_index=False)[['revenue', 'transactionId']]\
.agg({'revenue':'sum', 'transactionId':'count'})
visitors_revenue_A.columns = ['visitorId', 'sum_revenue', 'N_transaction']

# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª-–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ B
visitors_revenue_B = orders[orders['group'] == 'B'].groupby('visitorId', as_index=False)[['revenue', 'transactionId']]\
.agg({'revenue':'sum', 'transactionId':'count'})
visitors_revenue_B.columns = ['visitorId', 'sum_revenue', 'N_transaction']

display('visitors_revenue_A:', visitors_revenue_A.sort_values(by='N_transaction', ascending=False),\
        'visitors_revenue_B:', visitors_revenue_B.sort_values(by='N_transaction', ascending=False))

print('p-value = ', round(st.mannwhitneyu(visitors_revenue_B['sum_revenue'],visitors_revenue_A['sum_revenue'])[1], 5))
print('p-value = ', round(st.mannwhitneyu(visitors_revenue_B['sum_revenue']/visitors_revenue_B['N_transaction'],\
                                           visitors_revenue_A['sum_revenue']/visitors_revenue_A['N_transaction'])[1], 5))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B =','{0:.2%}'\
      .format(visitors_revenue_B['sum_revenue'].mean()/visitors_revenue_A['sum_revenue'].mean()- 1))



         
```

**–ü–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º –ø–æ–ª—É—á–∞–µ–º —Å–ª–µ–¥—É—é—â—É—é –æ—Ü–µ–Ω–∫—É —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –Ω–∞–ª–∏—á–∏—è —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –í –∏ –ê.**  

**–ü—Ä–∏ –ø–æ—Ä–æ–≥–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∞ = 0,05 –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–∫–ª–æ–Ω—è—Ç—å –Ω–µ—Ç. –í—ã–±–∏—Ä–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ B –∏ –ê –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç. –ü—Ä–∏ —ç—Ç–æ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ 27%, —á—Ç–æ –º–æ–∂–µ—Ç –æ–±—ä—è—Å–Ω—è—Ç—å—Å—è –Ω–∞–ª–∏—á–∏–µ–º –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.**  

**–ö—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –ø—Ä–∏–º–µ–Ω—è–ª–∏ —Å–Ω–∞—á–∞–ª–∞ –∫ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—è–º –≤—ã—Ä—É—á–µ–∫ "visitors_revenue_A[sum_revenue]", –∞ –∑–∞—Ç–µ–º –∫ –±–æ–ª–µ–µ —Ç–æ—á–Ω—ã–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—è–º —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ "visitors_revenue_A[sum_revenue]/visitors_revenue_–ê['N_transaction']". –ï—Å–ª–∏ –±—ã –≤—Å–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ –∏–º–µ–ª–∏ —Ç–æ–ª—å–∫–æ –ø–æ –æ–¥–Ω–æ–º—É –∑–∞–∫–∞–∑—É, —Ç–æ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –±—ã–ª–∏ –±—ã –∏–¥–µ–Ω—Ç–∏—á–Ω—ã. –ò —á–µ–º –±–æ–ª—å—à–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Å –∫–æ–ª-–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ >1, —Ç–µ–º —Å–∏–ª—å–Ω–µ–µ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è —ç—Ç–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏. –ü—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–∏—Ö —á–µ–∫–æ–≤ –∫—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –¥–∞–µ—Ç –±–æ–ª–µ–µ –≤—ã—Å–æ–∫—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å (82% –ø—Ä–æ—Ç–∏–≤ 72,5%) –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í.**

```python
# –≤—ã–±–µ—Ä–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–≤ –±–æ–ª—å—à–µ (perc1 = 95) –∏ —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–∞ –±–æ–ª—å—à–µ (perc2 = 95) –¥–æ–ª–ª
perc1 = 95
perc2 = 95
abnormal_visitors = pd.concat(
    [visitors_transactions[visitors_transactions['N_transactions'] > 
                           np.percentile(visitors_transactions['N_transactions'], perc1)]['visitorId'],
     visitors_revenue[visitors_revenue['sum_revenue'] > 
                      np.percentile(visitors_revenue['sum_revenue'], perc2)]['visitorId']
    ], axis=0).drop_duplicates().sort_values()

print(f'–ö–æ–ª-–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Å —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–≤ –±–æ–ª—å—à–µ {np.percentile(visitors_transactions["N_transactions"], 95)} \
–∏ —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–∞ –±–æ–ª—å—à–µ {round(np.percentile(visitors_revenue["sum_revenue"], 95))} –¥–æ–ª–ª - \
{abnormal_visitors.count()} –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–ª–∏ {round(abnormal_visitors.count()/orders["visitorId"].count()*100,1)}%\
 –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
```

```python
# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –±–µ–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –∞–Ω–æ–º–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ –≥—Ä—É–ø–ø–µ –ê –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏
sample_A_filtered = pd.concat(
    [
     visitors_transactions_A.query('visitorId not in @abnormal_visitors')
        ['N_transactions'],
     pd.Series(0,index=np.arange(N_zero_A), name='transactions')
    ], axis=0)



# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –ø–æ –≥—Ä—É–ø–ø–µ –í –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏  np.logical_not(visitors_transactions_A['visitorId'].isin(abnormal_visitors))
sample_B_filtered = pd.concat(
    [
     visitors_transactions_B[np.logical_not(visitors_transactions_B['visitorId'].isin(abnormal_visitors))]
        ['N_transactions'],
     pd.Series(0,index=np.arange(N_zero_B), name='transactions')
    ], axis=0)

display(sample_A, sample_B)

print('p-value = ', round(st.mannwhitneyu(sample_A_filtered, sample_B_filtered)[1], 5))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B =','{0:.2%}'
      .format(sample_B_filtered.mean()/sample_A_filtered.mean()- 1))
         
```

**–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø—Ä–∏ –ø–æ—Ä–æ–≥–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∞ = 0,05 –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–∏—Å–ª–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í –Ω–µ—Ç, - –æ—Ç–≤–µ—Ä–≥–∞–µ–º –∫–∞–∫ –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º, —Ç–∞–∫ –∏ –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º".

### –§–î. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–∞–º –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º. –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
orders.info()
print(len(orders['visitorId'].unique()), len(orders['transactionId'].unique()))
print(len(abnormal_visitors))

orders_filtered = orders.query('visitorId not in @abnormal_visitors')
orders_filtered.info()
# —Å—Ç—Ä–æ–∏–º –º–∞—Å—Å–∏–≤ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø
dates_groups_filtered = orders[['date', 'group']].drop_duplicates().sort_values(by=['date', 'group'])

# –∞–≥—Ä–µ–≥–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É "orders_filtered"
orders_filtered_agregg = dates_groups_filtered.apply(
    lambda x:orders_filtered[
        np.logical_and(orders_filtered['date'] <= x['date'], orders_filtered['group'] == x['group'])
                    ]
        .agg({'date':'max', 'group':'max', 'visitorId':'nunique', 'transactionId':'nunique', 'revenue':'sum'}), axis=1
                                    )\
        .sort_values(by=['date', 'group'])
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü–∞–º –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ —Ç–∞–±–ª–∏—Ü–µ "orders_filtered_agregg"
orders_filtered_agregg.columns = ['date', 'group', 'cum_N_buyers', 'cum_N_orders', 'cum_revenue']  

display('–û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–∞—è –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ orders_filtered_agregg:', orders_filtered_agregg.head())
orders_filtered_agregg.info()

# –∞–≥—Ä–µ–≥–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É "visitors"
visitors_agregg = dates_groups.apply(
    lambda x:visitors[
        np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])
                    ]
        .agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1
                                    )\
        .sort_values(by=['date', 'group'])
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü–∞–º –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ —Ç–∞–±–ª–∏—Ü–µ "visitors_agregg"
visitors_agregg.columns = ['date', 'group', 'cum_N_visitors']  

# –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
display(visitors_agregg.head())
visitors_agregg.info()

# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_filtered_agregg" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_filtered_agregg" –∏ "visitors_agregg"
orders_filtered_visitors_agregg = orders_filtered_agregg.merge(visitors_agregg, left_on=['date', 'group'], \
                                                               right_on=['date', 'group'], how='left')

# –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å
display(orders_filtered_visitors_agregg.head())
orders_filtered_visitors_agregg.info()

# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_filtered_visitors_agregg_A = orders_filtered_visitors_agregg[orders_filtered_visitors_agregg['group'] == 'A']
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_filtered_visitors_agregg_B = orders_filtered_visitors_agregg[orders_filtered_visitors_agregg['group'] == 'B']

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –ê
plt.plot(orders_filtered_visitors_agregg_A['date'], orders_filtered_visitors_agregg_A['cum_N_orders']/orders_filtered_visitors_agregg_A['cum_N_visitors'],\
         label='A')
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã –í
plt.plot(orders_filtered_visitors_agregg_B['date'], orders_filtered_visitors_agregg_B['cum_N_orders']/orders_filtered_visitors_agregg_B['cum_N_visitors'],\
         label='B')
plt.legend()
plt.title('–§–∏–ª—å—Ç—Ä.–î. –ì—Ä–∞—Ñ–∏–∫–∏ –∫—É–º—É–ª—è—Ç-–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=18, color='blue')
plt.ylabel('–§–∏–ª—å—Ç—Ä.–î. –ö—É–º—É–ª—è—Ç-–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.axhline(y=0.026, color='red', linestyle='dotted')
plt.axhline(y=0.022, color='green', linestyle='dotted')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.show()
```

**–ì—Ä–∞—Ñ–∏–∫ –≤ –æ–±—â–µ–º –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è. –ü—Ä–æ—Å–µ–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ (9-10%%) —É—Ä–æ–≤–Ω–∏ —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é—â–∏—Ö—Å—è –∑–Ω–∞—á–µ–Ω–∏–π. –†–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ —Å–æ—Ö—Ä–∞–Ω–∏–ª–æ—Å—å.**

### –§–î. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º.  –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_filtered_visitors_agregg_–ê–í" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_filtered_visitors_agregg_–ê" –∏ "orders_filtered_visitors_agregg_B"
orders_filtered_visitors_agregg_AB = orders_filtered_visitors_agregg_A.merge(orders_filtered_visitors_agregg_B, left_on='date', right_on='date', \
                                                           how='left', suffixes=['A','B'])

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A
plt.plot(orders_filtered_visitors_agregg_AB['date'], \
         (orders_filtered_visitors_agregg_AB['cum_N_ordersB']/orders_filtered_visitors_agregg_AB['cum_N_visitorsB'])\
         /(orders_filtered_visitors_agregg_AB['cum_N_ordersA']/orders_filtered_visitors_agregg_AB['cum_N_visitorsA']) - 1,\
         label='(B/A - 1)')
plt.legend()
plt.title('–§–∏–ª—å—Ç—Ä.–î. –ì—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç-–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç-–≥–æ –∫–æ–ª-–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –≥—Ä.B –∫ –≥—Ä.A', \
          fontsize=18, color='blue')
#plt.ylabel('–í—ã—Ä—É—á–∫–∞, $', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.axhline(y=0.18, color='green', linestyle='--')
plt.axhline(y=0, color='black', linestyle='-')
plt.show()
```

**–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ. –ì—Ä–∞—Ñ–∏–∫ –≤ –æ–±—â–µ–º –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è, –Ω–æ —Å—Ç–∞–ª –±–æ–ª–µ–µ "–∫–æ–ª–µ–±–∞—Ç–µ–ª—å–Ω—ã–º".  –†–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ —Å–æ—Ö—Ä–∞–Ω–∏–ª–æ—Å—å.**

### –§–î. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º. –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ –ê
visitors_revenue_A_filtered = orders.query('group == "A" & visitorId not in @abnormal_visitors')\
.groupby('visitorId', as_index=False)[['revenue', 'transactionId']]\
.agg({'revenue':'sum', 'transactionId':'count'})
visitors_revenue_A_filtered.columns = ['visitorId', 'sum_revenue', 'N_transaction']

# —Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–ª-–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –≤ –≥—Ä—É–ø–ø–µ B
visitors_revenue_B_filtered = orders.query('group == "B" & visitorId not in @abnormal_visitors')\
.groupby('visitorId', as_index=False)[['revenue', 'transactionId']]\
.agg({'revenue':'sum', 'transactionId':'count'})
visitors_revenue_B_filtered.columns = ['visitorId', 'sum_revenue', 'N_transaction']

display('visitors_revenue_A_filtered:', visitors_revenue_A_filtered.sort_values(by='N_transaction', ascending=False),'visitors_revenue_B_filtered:', visitors_revenue_B_filtered.sort_values(by='N_transaction', ascending=False))

print('p-value = ', round(st.mannwhitneyu(visitors_revenue_B_filtered['sum_revenue'],\
                                          visitors_revenue_A_filtered['sum_revenue'])[1], 5))
print('p-value = ', round(st.mannwhitneyu(visitors_revenue_B_filtered['sum_revenue']/\
                                          visitors_revenue_B_filtered['N_transaction'],\
                                           visitors_revenue_A_filtered['sum_revenue']/\
                                          visitors_revenue_A_filtered['N_transaction'])[1], 5))
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –∑–∞–∫–∞–∑–æ–≤ –≥—Ä—É–ø–ø—ã B =','{0:.2%}'\
      .format(visitors_revenue_B_filtered['sum_revenue'].mean()/visitors_revenue_A_filtered['sum_revenue'].mean()- 1))

```

**–ü–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ p-value —É–≤–µ–ª–∏—á–∏–ª–æ—Å—å –¥–æ 0,83632 (–±—ã–ª–æ 0,82146), —á—Ç–æ, –µ—â—ë –±–æ–ª–µ–µ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –í –∏ –ê. –ü—Ä–∏ —ç—Ç–æ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º —É–ø–∞–ª –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –¥–æ 0% (-0,9%)(–ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º –±—ã–ª 27%), —á—Ç–æ –º–æ–∂–Ω–æ –æ–±—ä—è—Å–Ω–∏—Ç—å –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –∞–Ω–æ–º–∞–ª—å–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞, –∫–æ—Ç–æ—Ä—ã–π –∫–∞–∫ —Ä–∞–∑ –ø—Ä–∏—Ö–æ–¥–∏–ª—Å—è –Ω–∞ –≥—Ä—É–ø–ø—É –í, —Å —Å—É–º–º–æ–π –±–æ–ª–µ–µ —á–µ–º –≤ 40 —Ä–∞–∑ –ø—Ä–µ–≤—ã—à–∞—é—â–µ–π –ø–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 95 –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è.**  

**–û—Ç–º–µ—Ç–∏–º, —á—Ç–æ –ø–æ—Å–ª–µ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ —É –≤—Å–µ—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –ø–æ 1 –∑–∞–∫–∞–∑—É –∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –≤—ã–±–æ—Ä–∫–∞ –ø–æ –≤—ã—Ä—É—á–∫–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–µ–Ω–Ω–∞ –≤—ã–±–æ—Ä–∫–µ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É —á–µ–∫—É, —á—Ç–æ –∏ –≤–∏–¥–∏–º –≤ –∑–Ω–∞—á–µ–Ω–∏—è—Ö p-value, –∫–æ—Ç–æ—Ä—ã–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–µ–Ω–Ω–æ —Ä–∞–≤–Ω—ã.**  

**–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø—Ä–∏ –ø–æ—Ä–æ–≥–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∞ = 0,05 –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ  –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ê –∏ –í –Ω–µ—Ç, - –ø—Ä–∏–Ω–∏–º–∞–µ–º –∫–∞–∫ –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º, —Ç–∞–∫ –∏ –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º".  
–í—ã–±–∏—Ä–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Ä–∞–∑–ª–∏—á–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–æ–≤ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ B –∏ –ê –Ω–µ—Ç.  
–ü—Ä–∏ —ç—Ç–æ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ -1%.**

### –§–î. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –ø–æ –≥—Ä—É–ø–ø–∞–º. –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
orders.info()
print(len(orders['visitorId'].unique()), len(orders['transactionId'].unique()))
print(len(abnormal_visitors))

orders_filtered = orders.query('visitorId not in @abnormal_visitors')
orders_filtered.info()
# —Å—Ç—Ä–æ–∏–º –º–∞—Å—Å–∏–≤ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø
dates_groups_filtered = orders[['date', 'group']].drop_duplicates().sort_values(by=['date', 'group'])

# –∞–≥—Ä–µ–≥–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É "orders_filtered"
orders_filtered_agregg = dates_groups_filtered.apply(
    lambda x:orders_filtered[
        np.logical_and(orders_filtered['date'] <= x['date'], orders_filtered['group'] == x['group'])
                    ]
        .agg({'date':'max', 'group':'max', 'visitorId':'nunique', 'transactionId':'nunique', 'revenue':'sum'}), axis=1
                                    )\
        .sort_values(by=['date', 'group'])
# –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü–∞–º –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ —Ç–∞–±–ª–∏—Ü–µ "orders_filtered_agregg"
orders_filtered_agregg.columns = ['date', 'group', 'cum_N_buyers', 'cum_N_orders', 'cum_revenue']  

display('–û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω–∞—è –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ orders_filtered_agregg:', orders_filtered_agregg.head())
orders_filtered_agregg.info()

# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_filtered_agregg_A = orders_filtered_agregg[orders_filtered_agregg['group'] == 'A']
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
orders_filtered_agregg_B = orders_filtered_agregg[orders_filtered_agregg['group'] == 'B']

print('–ö–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ orders_filtered_agregg_A:', orders_agregg_A['group'].count())
print('–ö–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ orders_filtered_agregg_B:', orders_agregg_B['group'].count())

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –ê
plt.plot(orders_filtered_agregg_A['date'], orders_filtered_agregg_A['cum_revenue'], label='A')
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –í
plt.plot(orders_filtered_agregg_B['date'], orders_filtered_agregg_B['cum_revenue'], label='B')
# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_agregg_–ê–í" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_agregg_–ê" –∏ "orders_agregg_B"
orders_filtered_agregg_AB = orders_filtered_agregg_A.merge(orders_filtered_agregg_B, left_on='date', right_on='date', \
                                                           how='left', suffixes=['A','B'])
plt.plot(orders_filtered_agregg_B['date'], orders_filtered_agregg_AB['cum_revenueB']\
         - orders_filtered_agregg_AB['cum_revenueA'], label='B-–ê')
plt.legend()
plt.title('–§–∏–ª—å—Ç—Ä.–î. –ì—Ä–∞—Ñ–∏–∫–∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=18, color='blue')
plt.ylabel('–í—ã—Ä—É—á–∫–∞, $*100000', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.show()
```

**–°—Ç—É–ø–µ–Ω—å–∫–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≥—Ä—É–ø–ø—ã –í —É—Å—Ç—Ä–∞–Ω–µ–Ω–∞, –∞–Ω–æ–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–∞–∫–∞–∑–∞ –æ—Ç 2019-08-19 —É–¥–∞–ª–µ–Ω–æ. –ò —Ç–µ–ø–µ—Ä—å —á–µ—Ç–∫–æ –≤–∏–¥–Ω–æ, –≥—Ä–∞—Ñ–∏–∫–∏ –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ –≤–æ–∑—Ä–∞—Å—Ç–∞—é—Ç, –ø—Ä–∏ —ç—Ç–æ–º –≥—Ä–∞—Ñ–∏–∫ –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –í –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –≤—Å–µ-—Ç–∞–∫–∏ –Ω–µ–º–Ω–æ–≥–æ –±—ã—Å—Ç—Ä–µ–µ. –ì—Ä–∞—Ñ–∏–∫ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –≤—ã—Ä—É—á–µ–∫ –≥—Ä—É–ø–ø –í –∏ –ê –∫–æ–ª–µ–±–ª–µ—Ç—Å—è, –Ω–æ –∏–º–µ–µ—Ç —á–µ—Ç–∫–∏–π –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–∏–π —Ç—Ä–µ–Ω–¥. –í—ã—Ä—É—á–∫–∞ –≥—Ä—É–ø–ø—ã –í –ø—Ä–µ–≤—ã—à–∞–µ—Ç –≤—ã—Ä—É—á–∫—É –≥—Ä—É–ø–ø—ã –ê –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–≤–Ω–æ.**

### –§–î. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º –ø–æ –≥—Ä—É–ø–ø–∞–º.  –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã –ê
plt.plot(orders_filtered_agregg_A['date'], orders_filtered_agregg_A['cum_revenue']/orders_filtered_agregg_A['cum_N_orders'], label='A')
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã –í
plt.plot(orders_filtered_agregg_B['date'], orders_filtered_agregg_B['cum_revenue']/orders_filtered_agregg_B['cum_N_orders'], label='B')
plt.legend()
plt.title('–§–∏–ª—å—Ç—Ä.–î. –ì—Ä–∞—Ñ–∏–∫–∏ –∫—É–º—É–ª—è—Ç-–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=18, color='blue')
plt.ylabel('–í—ã—Ä—É—á–∫–∞, $', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.show()
```

**–°—Ç—É–ø–µ–Ω—å–∫–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≥—Ä—É–ø–ø—ã –í —É—Å—Ç—Ä–∞–Ω–µ–Ω–∞, –∞–Ω–æ–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–∞–∫–∞–∑–∞ –æ—Ç 2019-08-19 —É–¥–∞–ª–µ–Ω–æ. –ò —Ç–µ–ø–µ—Ä—å —á–µ—Ç–∫–æ –≤–∏–¥–Ω–æ, —á—Ç–æ –≥—Ä–∞—Ñ–∏–∫–∏ –∫–æ–ª–µ–±–ª—é—Ç—Å—è, –∫–∞–∂–¥—ã–π –ø–æ-—Å–≤–æ–µ–º—É, –∏–º–µ—é—Ç –∑–∞—Ç—É—Ö–∞—é—â–∏–π –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â–∏–π —Ç—Ä–µ–Ω–¥ –∏ —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é –∫ —Å–±–ª–∏–∂–µ–Ω–∏—é –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º. –ù–∏–∫–∞–∫–æ–≥–æ –∑–Ω–∞—á–∏–º–æ–≥–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ –≤–µ–ª–∏—á–∏–Ω–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º –Ω–µ—Ç.**

### –§–î. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A –ø–æ "—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º.  –í—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
# —Å–æ–∑–¥–∞–¥–∏–º –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É "orders_agregg_–ê–í" –Ω–∞ –æ—Å–Ω–æ–≤–µ "orders_agregg_–ê" –∏ "orders_agregg_B"
orders_filtered_agregg_AB = orders_filtered_agregg_A.merge(orders_filtered_agregg_B, left_on='date', right_on='date', \
                                                           how='left', suffixes=['A','B'])

plt.figure(figsize=(14,7))
plt.grid()
# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã B –∫ –≥—Ä—É–ø–ø–µ A
plt.plot(orders_filtered_agregg_AB['date'], (orders_filtered_agregg_AB['cum_revenueB']/orders_filtered_agregg_AB['cum_N_ordersB'])\
         /(orders_filtered_agregg_AB['cum_revenueA']/orders_filtered_agregg_AB['cum_N_ordersA']) - 1, label='(B/A - 1)')
plt.legend()
plt.title('–§–∏–ª—å—Ç—Ä.–î. –ì—Ä–∞—Ñ–∏–∫ –æ—Ç–Ω–æ—Å–∏—Ç-–≥–æ –∏–∑–º–µ–Ω-—è –∫—É–º—É–ª—è—Ç-–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä.B –∫ –≥—Ä.A', fontsize=18, color='blue')
#plt.ylabel('–í—ã—Ä—É—á–∫–∞, $', fontsize=15, color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('–î–∞—Ç—ã', fontsize=15, color='blue')
plt.axhline(y=0.2, color='green', linestyle='--')
plt.axhline(y=0, color='black', linestyle='-')
plt.show()
```

**–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø—Ä–µ–¥—ã–¥—É—â–µ–º—É –≥—Ä–∞—Ñ–∏–∫—É. –°—Ç—É–ø–µ–Ω—å–∫–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ —É—Å—Ç—Ä–∞–Ω–µ–Ω–∞, –∞–Ω–æ–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–∞–∫–∞–∑–∞ –æ—Ç 2019-08-19 —É–¥–∞–ª–µ–Ω–æ. –ò —Ç–µ–ø–µ—Ä—å —á–µ—Ç–∫–æ –≤–∏–¥–Ω–æ, –≥—Ä–∞—Ñ–∏–∫ –∫–æ–ª–µ–±–ª–µ—Ç—Å—è, –∏–º–µ—é—Ç –∑–∞—Ç—É—Ö–∞—é—â–∏–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∏ —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é –∫ —Å–±–ª–∏–∂–µ–Ω–∏—é —Å –Ω—É–ª–µ–≤–æ–π –æ—Å—å—é, –æ—Å—å—é –∞–±—Å—Ü–∏—Å—Å. –ù–∏–∫–∞–∫–æ–≥–æ –∑–Ω–∞—á–∏–º–æ–≥–æ —Ä–∞–∑–ª–∏—á–∏—è –≤ –≤–µ–ª–∏—á–∏–Ω–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º –Ω–µ—Ç.**

### –ü—Ä–∏–Ω—è—Ç–∏–µ —Ä–µ—à–µ–Ω–∏—è –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º —Ç–µ—Å—Ç–∞ –∏ –æ–±—ä—è—Å–Ω–µ–Ω–∏–µ –µ–≥–æ.

**–ü–æ–¥–≤–µ–¥–µ–º –∏—Ç–æ–≥–∏,**
